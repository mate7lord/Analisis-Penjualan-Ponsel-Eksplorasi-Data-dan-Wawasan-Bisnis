## Analisis Data Eksploratif (EDA)

Sekarang kita akan melakukan beberapa visualisasi data untuk mendapatkan wawasan lebih dalam. Jalankan sel-sel berikut untuk melihat berbagai visualisasi.


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Distribusi Harga
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], kde=True)
plt.title('Distribusi Harga')
plt.show()
```

```python
# Distribusi Usia Pelanggan
plt.figure(figsize=(10, 6))
sns.histplot(data['CustomerAge'], kde=True)
plt.title('Distribusi Usia Pelanggan')
plt.show()
```

```python
# Penjualan Seiring Waktu
plt.figure(figsize=(10, 6))
data.set_index('Date').resample('M')['TotalRevenue'].sum().plot()
plt.title('Total Pendapatan Seiring Waktu')
plt.show()
```

```python
# Ambil 10 merek dengan penjualan unit tertinggi
top_brands = data.groupby('Brand')['UnitsSold'].sum().nlargest(10).index
filtered_data = data[data['Brand'].isin(top_brands)]
```

```python
# Unit Terjual per Merek dengan merek teratas
plt.figure(figsize=(14, 8))
sns.boxplot(x='Brand', y='UnitsSold', data=filtered_data, palette='Set3')
plt.title('Unit Terjual per Merek')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Brand')
plt.ylabel('Units Sold')
plt.show()
```

```python
# Total Pendapatan berdasarkan Metode Pembayaran
plt.figure(figsize=(10, 6))
sns.boxplot(x='PaymentMethod', y='TotalRevenue', data=data, palette='Set3')
plt.title('Total Pendapatan berdasarkan Metode Pembayaran')
plt.show()
```

# Rangkuman temuan dan wawasan yang dapat ditindaklanjuti
1. Harga Ponsel: Analisis menunjukkan bahwa rata-rata harga ponsel yang terjual adalah sekitar $807.51. Ini memberikan gambaran tentang kisaran harga ponsel yang populer di pasar.
2. Usia Pelanggan: Rata-rata usia pelanggan adalah 41 tahun, yang memberikan wawasan tentang demografi pelanggan.
3. Penjualan Merek: Penjualan terlihat cukup merata di berbagai merek tanpa ada outlier yang signifikan, menunjukkan bahwa tidak ada merek tertentu yang mendominasi pasar secara ekstrem.
4. Metode Pembayaran: Sebagian besar pelanggan lebih memilih metode pembayaran online, diikuti oleh pembayaran dengan kartu kredit. Ini penting untuk strategi pembayaran dan penawaran promosi.
5. Tren Pendapatan Bulanan: Pendapatan menunjukkan tren bulanan yang dapat digunakan untuk perencanaan strategi pemasaran dan penjualan, seperti kapan waktu yang tepat untuk melakukan promosi besar-besaran.
